```python
#Fog Stats


fog_dist = {}
df['Fog_avg'] = 0

for idx, x in enumerate(df['principal_strategies']):
    list_fog = []
    
    for s in nltk.sent_tokenize(x):
        fog = textstat.gunning_fog(s)
        list_fog.append(fog)
        
    fog_dist[idx] = list_fog

    try:
        df['Fog_avg'][idx] = mean(list_fog)
    except:
        df['Fog_avg'][idx] = 0
        
df['Fog_dist'] = pd.Series(fog_dist)
```

```python
#ARI Stats


ari_dist = {}
df['Ari_avg'] = 0

for idx, x in enumerate(df['principal_strategies']):
    list_ari = []
    
    for s in nltk.sent_tokenize(x):
        ari = textstat.automated_readability_index(s)
        list_ari.append(ari)

    ari_dist[idx] = list_ari

    try:
        df['Ari_avg'][idx] = mean(list_ari)
    except:
        df['Ari_avg'][idx] = 0
        
df['Ari_dist'] = pd.Series(ari_dist)
```

```python
#Passivity
#Get the verb ratio and the sentence ratio as means and dists.


Pass_verb_avg= {}
Pass_sent_avg = {}
df['Sent_count_avg'] = 0
Pass_verb_dist = {}
Pass_sent_dist = {}

beforms = ['be', 'was', 'were', 'been', 'being', 'am', 'is', 'are']

#Document Loop
for idx, x in enumerate(df['principal_strategies']): 
    ratio_list = []
    passe_list = []
    sent_count = len(nltk.sent_tokenize(x))
    
    #Sentence Loop
    for sent in nltk.sent_tokenize(x):
        #print(sent)
        doc = nlp(sent)
        active_list = []
        passive_list = []
        
        for token in doc:  
            #print(token)
            #print(token.text, token.lemma_, token.pos_, token.tag_,)
            if token.lemma_ in beforms:
                passive_list.append(token.text)
            if token.pos_ == 'VERB':
                active_list.append(token.text) 
                
        #verb ratio per doc
        if len(active_list) == 0:
            ratio_list.append(1)
        else:
            ratio_list.append(len(passive_list)/(len(active_list)))
        #print('Pass_Act: ', ratio_list) 
        
        #sentence ratio per doc
        if sent_count == 0:
            passe_list.append(0)
        else:
            passe_list.append(len(passive_list)/sent_count)
        #print('Pass_Per: ', passe)
    
    Pass_verb_avg[idx] = mean(ratio_list)
    Pass_sent_avg[idx] = mean(passe_list)
    df['Sent_count_avg'][idx] = sent_count
    Pass_verb_dist[idx] = ratio_list
    Pass_sent_dist[idx] = passe_list
    

df['Pass_verb_avg'] = pd.Series(Pass_verb_avg)
df['Pass_sent_avg'] = pd.Series(Pass_sent_avg)
df['Pass_verb_dist'] = pd.Series(Pass_verb_dist)
df['Pass_sent_dist'] = pd.Series(Pass_sent_dist)
```
